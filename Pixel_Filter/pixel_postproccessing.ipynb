# This notebook shows what each filter does

```python
from filters import*
from PIL import Image  # for display
```

```python
# Load and display unfiltered image
path = "sample.png"
output_path = "filtered_sample.png"
img = Image.open(path)
img_array = cv2.imread(path)

img
```

```python
# Scaling

new_modified_image = upscale_nearest_neighbor(img_array, 6) # filter and write
cv2.imwrite(output_path, new_modified_image)

Image.open(output_path) # display
```

```python
# Luminance manipulation

new_modified_image = adjust_rgb(img_array, 1.5,0.4,0.3) # weighs red, green, blue values, if r=g=b then only adjusts luminance
cv2.imwrite(output_path, new_modified_image)

Image.open(output_path) # display
```

```python
# very simple box blur

img_array = cv2.imread(path)
new_img_array = np.zeros(img_array.shape)
new_modified_image = apply_box_blur_3(new_img_array, img_array)
cv2.imwrite(output_path, new_modified_image)
Image.open(output_path)
```

```python
# Pixel Screen Version 1

img_array = cv2.imread(path)
new_modified_image = rgb_effect(adjust_rgb(img_array,2,2,2), 3)
cv2.imwrite(output_path, new_modified_image)
Image.open(output_path)
```

```python
# Dotted version 1

new_modified_image = upscale_square_dots(img_array, 10,3) # 10 is scale, 3 is dot width
cv2.imwrite(output_path, new_modified_image)

Image.open(output_path) # display
```
